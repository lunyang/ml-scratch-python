### 逻辑斯特回归案例

#### 生成数据

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(42)
num_observations = 5000
x1 = np.random.multivariate_normal([0,0], [[1, .75], [.75, 1]], num_observations)
x2 = np.random.multivariate_normal([1,4], [[1, .75], [.75, 1]], num_observations)

simulated_sepatableish_features = np.vstack((x1, x2)).astype(np.float32)
simulated_labels = np.hstack((np.zeros(num_observations),
                             np.ones(num_observations)))
```

看下数据长什么样

```python
plt.figure(figsize=(10, 8))
plt.scatter(simulated_sepatableish_features[:,0], simulated_sepatableish_features[:,1],
           c = simulated_labels, alpha = .4)
```

### 定义链接函数

logistic回归模型中使用的链接函数是sigmoid函数，很容易定义

```python
def sigmoid(scores):
    return 1/(1 + np.exp(-scores))
```

### 最大化似然函数

#### 计算Log-Likelihood

$$L(\beta) = \sum \limits_i{y_i \beta ^T x_i - log(1+e^{\beta ^ T x_i})}$$  
其中$y$是目标变量，取值为$\{0,1\}$, $x_i$是一个样本点， $\beta$为权重矢量.

```python
def log_likeihood(feature, target, weights):
    scores = np.dot(feature, weights)
    ll = np.sum(target*scores - np.log(1 + np.exp(scores))) # log likelihood
    return ll
```

### 计算梯度

梯度为：  
$$\Delta L(\beta) = X ^ T(Y-Predictions) $$

### 构建逻辑斯特回归函数

```python
def logistic_regression(features, target, num_steps, learning_rate, add_intercept = False):
    if add_intercept:
        intercept = np.ones((features.shape[0],1)) # 截距特征为1
        features = np.hstack((intercept, features))
    weights = np.zeros(features.shape[1]) # 初始化权重为0
    
    for step in range(num_steps):
        scores = np.dot(features, weights)
        predictions = sigmoid(scores)
        
        # 梯度下降法更新权重
        output_error_signal = target - predictions
        gradient = np.dot(features.T, output_error_signal)
        weights += learning_rate * gradient
        
        # 打印似然函数
        if step % 10000 ==0:
            print(log_likeihood(features, target, weights))
    return weights
    
```

测试算法

```python
weight = logistic_regression(simulated_sepatableish_features, simulated_labels, num_steps = 300000, learning_rate = 5e-5, add_intercept = True)
```

### 与sklearn中的LR算法对比

sklearn中的LogisticRegression使用了L2正则化， 这里设置C=1e15

```python
from sklearn.linear_model import  LogisticRegression
clf = LogisticRegression(fit_intercept=True, C = 1e15)
clf.fit(simulated_sepatableish_features, simulated_labels)
```

```python
print(clf.intercept_, clf.coef_)
print(weight)
```

可以看到我们的结果和sklearn的结果很接近了。

### 预测精确度

将得到的权重输入到sigmoid函数即可得到预测值，0,1之间

```python
data_with_intercept = np.hstack((np.ones((simulated_sepatableish_features.shape[0],1)), simulated_sepatableish_features))
final_scores = np.dot(data_with_intercept, weight)
pred_probs = np.round(sigmoid(final_scores))

print('Accracy from scratch:{0}'.format((pred_probs == simulated_labels).sum() / len(pred_probs)))
print('Accracy from sklearn: {0}'.format(clf.score(simulated_sepatableish_features, simulated_labels)))
```

需要注意的是我们并没有划分数据集为训练集/测试集，所有结果仅供演示学习使用，下面可视化下分类结果

```python
plt.figure(figsize = (12, 8))
plt.scatter(simulated_sepatableish_features[:, 0], simulated_sepatableish_features[:, 1],
            c = pred_probs == simulated_labels - 1, alpha = .8, s = 50)
```

### 总结

自行实现的算法没有考虑正则项，其它的优化算法以及多分类情况，差距最大最明显的是sklearn的**速度**直接秒杀自己实现的。
